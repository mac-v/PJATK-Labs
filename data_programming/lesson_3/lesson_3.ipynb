# Task 1.

```python
import pandas as pd

df = pd.read_csv('PAD_03_PD.csv', sep=";")
country_counts = df['Country'].value_counts()
country_counts
#df.groupby('Country').sum()
```

# Task 2

```python
def count_goods(row):
    return row['owns_car'] + row['owns_TV'] + row['owns_house'] + row['owns_Phone']
    
df['owned_goods'] = df.apply(count_goods, axis=1)
df
```

```python
df['owned_goods'] = df[['owns_car', 'owns_TV', 'owns_house', 'owns_Phone']].sum(axis=1)
df
```

# Task 3

```python
df_gender_avg = df[['gender', 'owned_goods']].groupby('gender').mean().round(2).reset_index()
df_gender_avg
```

# Task 4

```python
df = df[['Country', 'owned_goods','Age']].groupby('Country').agg(
    avg_goods=("owned_goods", "mean"),
    min_age=("Age", "min")
).round(2).reset_index()
df
```
