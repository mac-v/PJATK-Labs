```python
products_list = {}

for page in range(2, 32):
    products_info = driver.find_elements(By.CLASS_NAME, 'product-info')
    driver.maximize_window()
    driver.execute_script('window.scroll(0, document.body.scrollHeight-2500)')

    for i, element in enumerate(products_info):
        if element.text != '':
            item = {}
            item['price'] = element.find_element(By.CLASS_NAME,'btn.btn-block.btn-buy').find_element(By.CLASS_NAME,'product-price').text
            # item['price'] = element.find_element(By.XPATH, '/button/div/div/div').text
            product_name = element.find_element(By.CLASS_NAME, 'product-name')
            product_link = product_name.get_attribute('href')
            item['description'] = element.text
            item['item_page_id'] = f'{page-1}_{i}'
            # print(i, element.text)
            products_list[product_link] = item
    
    # NExt page clicker
    # driver.find_element(By.ID, 'p-next-bottom').find_element(By.CLASS_NAME, 'page-link').click()
    #Next page iterator 
    driver.get(f"https://www.coffeedesk.pl/kawa/?p={page}&order=data-palenia")
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess
import json

class JsonWriterPipeline(object):
    def open_spider(self, spider):
        self.file = open('quoteresult.json', 'w')

    def close_spider(self, spider):
        self.file.close()

    def process_item(self, item, spider):
        item_json = json.dumps(dict(item)) + '\n'
        self.file.write(item_json)

        return item
    

class MySpider(scrapy.Spider):
    name = 'mySpider'
    start_urls = [f'https://www.otodom.pl/pl/wyniki/sprzedaz/mieszkanie/cala-polska?viewType=listing&page={i}' for i in range(1, 400)]

    custom_settings = {
        'LOG_LEVEL': 'INFO',
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 2},
        'REQUEST_FINGERPRINTER_IMPLEMENTATION': '2.7',
        'FEEDS': {'quoteresult_1.json': {'format': 'json', 'overwrite': True}} 
        
    }
    
    def parse(self, response):
        listings = response.css('ul.css-rqwdxd.e127mklk0 > li > article')
        for listing in listings:
            yield self._parse_listig(listing, response)
        
    
    def _parse_listig(self, listing, response):

        item = {
            'link': 'N/A',
            'title':  'N/A',
            'Localisation':  'N/A',
            'price':  'N/A',
            'room_count':  'N/A',
            'space_sm': 'N/A',
            'floor':  'N/A',
            'representative': 'N/A'
        }

        # Extract the link
        item['link'] = listing.css('a[data-cy="listing-item-link"]::attr(href)').get(default='N/A')

        # Extract title
        item['title'] = listing.css('p[data-cy="listing-item-title"]::text').get(default='N/A')

        # Extract localisation
        item['localisation'] = listing.css('p.css-42r2ms::text').get(default='N/A')

        # Extract price and additional fee
        item['price'] = listing.css('span.css-2bt9f1::text').get(default='N/A').replace('\u00a0', ' ')

        # Extract rooms, space, and floor information from details
        parent_element = listing.css('dl.css-12dsp7a')
        for dt, dd in zip(parent_element.css('dt::text'), parent_element.css('dd::text')):
            label = dt.get()
            value = dd.get()
            match label:
                case 'Liczba pokoi':
                    item['room_count'] = value
                case 'Powierzchnia':
                    item['space_sm'] = value
                case 'Cena za metr kwadratowy':
                    item['floor'] = value

                
        # Extract representative (e.g., "Oferta prywatna" or other info)
        item['representative'] = listing.css('div.css-zmiifc::text').get(default='N/A')

        return item


process = CrawlerProcess({
    'USER_AGENT':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36',
    'REQUEST_FINGERPRINTER_IMPLEMENTATION': '2.7',

})

process.crawl(MySpider)
process.start()     
```
