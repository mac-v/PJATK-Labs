# Select sort

```python
def swap_elements(elements, first_index, second_index):
    temp = elements[first_index]
    elements[first_index] = elements[second_index]
    elements[second_index] = temp
    return elements


def sort(elements):
    i = 0
    min_index = 0
    compares_count = 0
    print("Initial elements:", elements)
    print("Initial minimal index:", min_index)
    while i < len(elements) - 1:
        min_index = i
        for index, element in enumerate(elements[i + 1:], start=i + 1):
            compares_count += 1
            print("Traversing: ", elements[i + 1:])
            if elements[min_index] > element:
                print("Comparing: ", elements[min_index], element)
                min_index = index
                print("Min index is: ", min_index)

        print("Swapping: ", min_index, i)
        elements = swap_elements(elements, min_index, i)
        print("After swapping:", elements)
        i += 1

    print("Compare count: ", compares_count)
    print("Ordered elements: ", elements)

elements = [5,3,1,8,7,4,2,6]
sort(elements)
```
