```python
import numpy as np

def cosine_similarity(A, B):
    dot_product = np.dot(A, B)
    norm_A = np.linalg.norm(A)
    norm_B = np.linalg.norm(B)
    cosine_sim = dot_product / (norm_A * norm_B)
    return cosine_sim

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

similarity = cosine_similarity(A, B)
print("Cosine Similarity:", similarity)
```

```python
def process_csv(file_path):
    df = pd.read_csv(file_path, header=None)
    return df.to_numpy()

matrix_a = process_csv('Zadanie_3_macierz_B.csv')
matrix_b = process_csv('Zadanie_3_macierz_B.csv')

def calculate_cosine_similarity_matrix(A, B):
    similarity_matrix = np.zeros((A.shape[0], B.shape[0]))

    for i in range(A.shape[0]):
        for j in range(B.shape[0]):
            vector_a = A[i, :]
            vector_b = B[j, :]

            dot_product = np.dot(vector_a, vector_b)

            norm_a = np.sqrt(np.sum(vector_a ** 2))
            norm_b = np.sqrt(np.sum(vector_b ** 2))
            similarity_matrix[i, j] = dot_product / (norm_a * norm_b)

    return similarity_matrix

result = calculate_cosine_similarity_matrix(matrix_a, matrix_b)

print("Cosine similarity matrix:")
print(np.round(result, 4))
```

```python
pd.set_option('display.max_rows', None)
df = pd.read_csv("Zadanie_4.csv", delimiter=";")
df
```

```python
import pandas as pd

df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d.%m.%Y %H:%M')
df.sort_values(by=['DoctorID', 'DateTime'], inplace=True)

df['Time_Diff'] = df.groupby(['DoctorID'])['DateTime'].diff().dt.seconds.div(60).fillna(1)

df['Group'] = (df['Time_Diff'] > 1).cumsum()

df = df.groupby(['DoctorID', 'Type', 'City', 'Group']).agg(
    Start=('DateTime', 'first'),
    End=('DateTime', 'last')
).reset_index()

print(df)
df.info()

```

```python
df.to_csv('doctor_availability', index=False)
```
